```python
import pandas as pd

df = pd.read_csv(r"C:\Users\jyoti\OneDrive\Desktop\New folder (3)\Untitled spreadsheet - Age_Input.csv")

pivot = (
    df.groupby(["Age", "Keyword"])
      .size()
      .unstack(fill_value=0)
)


pivot = pivot[["Intender", "Owner", "Rejector"]]


pivot["Grand Total"] = pivot.sum(axis=1)

age_order = ["Below 25", "25-34", "35-45", "46-50", "Above 50"]
pivot = pivot.reindex(age_order)


totals = pivot.sum()


pct = pivot.div(totals)


for col in pct.columns:
    pct[col] = (pct[col] * 100).round().astype(int).astype(str) + "%"


final = pd.concat([pivot, pct], axis=1)


final = final.reset_index()
final.rename(columns={"index": "Age"}, inplace=True)


totals_row = [""] + list(totals) + list(totals)
final.loc[len(final)] = totals_row


final.to_csv("Age_Output.csv", index=False)


```
